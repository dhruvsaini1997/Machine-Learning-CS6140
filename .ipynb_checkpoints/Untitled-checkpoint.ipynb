# IQR Method of Outlier Detection

Q1 is the first quartile of the data, i.e., to say 25% of the data lies between minimum and Q1.
Q3 is the third quartile of the data, i.e., to say 75% of the data lies between minimum and Q3.
The difference between Q3 and Q1 is called the Inter-Quartile Range or IQR.

**IQR = Q3 - Q1**

To detect the outliers using this method, we define a new range, letâ€™s call it decision range, and any data point lying outside this range is considered as outlier and is accordingly dealt with. The range is as given below:

**Lower Bound: (Q1 - 1.5 * IQR)**
**Upper Bound: (Q3 + 1.5 * IQR)**

Any data point less than the Lower Bound or more than the Upper Bound is considered as an **outlier**.


The choice of using **1.5 times the Interquartile Range (IQR)** is a common convention in statistics for identifying outliers in a dataset. This method is part of the Tukey's fences approach, developed by mathematician John Tukey. It's important to note that 1.5 times the IQR is not a strict rule but rather a widely accepted guideline.

Here's why 1.5 times the IQR is commonly used:

1. **Balance between sensitivity and robustness**: The choice of 1.5 times the IQR strikes a balance between being sensitive enough to detect potential outliers while still being robust enough to handle small variations and random fluctuations in the data.

2. **Reasonable threshold**: Using 1.5 times the IQR provides a reasonable threshold for identifying outliers. Values that fall beyond this threshold are considered potential outliers.

3. **Historical convention**: Over time, the 1.5 IQR rule has become widely adopted and taught in statistics courses, leading to its continued usage.

4. **Not too strict**: Some datasets may naturally have a spread of values, and being too strict with the threshold (e.g., using a larger multiplier) might lead to overlooking important information or potential anomalies.

That being said, the choice of the threshold can vary based on the specific context and the data being analyzed. In some cases, more or less strict thresholds may be used, such as 1.3 times or 2 times the IQR, depending on the specific analysis and domain expertise. It's essential to consider the characteristics of the dataset and the goals of the analysis when deciding on an appropriate threshold for outlier detection.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# given data
data = [11, 3, 8, 10, 12, 5, 1, 50]
sorted_data = sorted(data)

#len and index calc.

data_len = len(sorted_data)
part_1_index = data_len // 4
part_3_index = 3 * (data_len // 4)


# Calculate the first quartile (Q1)
quartile_1 = sorted_data[part_1_index] if data_len % 4 == 0 else (sorted_data[part_1_index] + sorted_data[part_1_index + 1]) / 2


# Calculate the third quartile (Q3)
quartile_3 = sorted_data[part_3_index] if data_len % 4 == 0 else (sorted_data[part_3_index] + sorted_data[part_3_index + 1]) / 2

print("Q1:", quartile_1)
print("Q3:", quartile_3)

# Calculate the Interquartile Range (IQR)
iqr = quartile_3 - quartile_1

# Calculate the lower and upper bounds for outliers
lower_bound = quartile_1 - 1.5 * iqr
print("lower_bound ", lower_bound)
upper_bound = quartile_3 + 1.5 * iqr
print("upper_bound ", upper_bound)

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers in data: " + ', '.join(map(str, outliers)))

# Create a box plot
plt.boxplot(data)
plt.title("Box Plot of Data")
plt.ylabel("Values")
plt.show()
```

##  Z-score Method for Outlier Detection

The Z-score, also known as the **standard score**, is a statistical measure that quantifies how many standard deviations a data point is from the mean of a dataset. It is used to understand how extreme or unusual a particular data point known as **outliers** is relative to the rest of the data. The formula for calculating the Z-score for an individual data point, x, in a dataset with mean (Î¼) and standard deviation (Ïƒ) is given by:

**Z = (x - Î¼) / Ïƒ**

Here's what each component of the formula represents:

- x: The value of the data point you want to calculate the Z-score for.
- Î¼: The mean of the dataset.
- Ïƒ: The standard deviation of the dataset.

The resulting Z-score tells you the number of standard deviations a data point is above or below the mean. A positive Z-score indicates that the data point is above the mean, while a negative Z-score indicates that it is below the mean.

Interpreting Z-scores:

- Z-score = 0: The data point is equal to the mean.
- Z-score > 0: The data point is above the mean by a certain number of standard deviations.
- Z-score < 0: The data point is below the mean by a certain number of standard deviations.
- The magnitude of the Z-score indicates how far the data point is from the mean in terms of standard deviations.

For example, if a data point has a Z-score of +2.0, it means that the value is two standard deviations above the mean. If it has a Z-score of -1.5, it means that the value is 1.5 standard deviations below the mean.

Z-scores are commonly used in statistics and data analysis to identify outliers and to compare values from different datasets with different scales and distributions. A Z-score outside the range of -3 to +3 is often considered unusual and potentially indicative of an outlier.

# Univariate Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# given data
data = [6, 3, 9, 6, 9, 20, 3, 10, 3, 50, 6, 5, 9, 9, 3, 6, 3]

# Calculate the mean and standard deviation of the data
mean = np.mean(data)
print("Mean: ",mean)
std_dev = np.std(data)
print("Standard Deviation: ",std_dev)

# Calculate the Z-score for each data point
z_scores = [(x - mean) / std_dev for x in data]
print("Z score matrix:")
print(z_scores)
# Set the threshold for outliers (Z-score greater than 3 or less than -3)
threshold = .9

# Identify outliers
outliers = [data[i] for i, z_score in enumerate(z_scores) if abs(z_score) > threshold]

print("Outliers:", outliers)

# Create a box plot
plt.boxplot(data)
plt.title("Box Plot of Data")
plt.ylabel("Values")
plt.show()
```

# Multivariate Example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load data from the CSV file
data_df = pd.read_csv('./datasets/Q1_dataset.csv')

# Extract the "SalePrice" column for analysis
sale_price = data_df["SalePrice"]

# Calculate z-scores for outlier detection
z_scores = np.abs(stats.zscore(sale_price))

# Define a threshold for z-score to mark outliers
z_score_threshold = 3

# Identify outlier indices
outlier_indices = np.where(z_scores > z_score_threshold)[0]

# Create a scatter plot to visualize outliers
plt.scatter(range(len(sale_price)), sale_price, label='SalePrice', color='b')
plt.scatter(outlier_indices, sale_price.iloc[outlier_indices], label='Outlier', color='r', marker='x')
plt.xlabel('Index')
plt.ylabel('SalePrice')
plt.title('Scatter Plot of SalePrice')
plt.legend()
plt.show()

# Create a histogram to visualize the distribution of SalePrice with outliers marked
plt.hist(sale_price, bins=20, edgecolor='black')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.title('Histogram of SalePrice')

# Mark outliers on the histogram
plt.axvline(sale_price.iloc[outlier_indices].min(), color='r', linestyle='dashed', linewidth=2, label='Outliers')
plt.axvline(sale_price.iloc[outlier_indices].max(), color='r', linestyle='dashed', linewidth=2)
plt.legend()
plt.show()
```
